```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Lectura de los datos
df = pd.read_csv('../data/processed/datos_integrados.csv')
df.head(5)
```

```python
resumen  = df.info()
```

```python
print(df.loc[1])
```

```python
categorical_cols = df.select_dtypes(include=["object"]).columns.drop("falta_pago")

print("\nDistribución de las variables categóricas del dataset de créditos:")

for col in categorical_cols:
    plt.figure(figsize=(8, 4))
    order = df[col].value_counts().index
    sns.countplot(y=col, data=df, order=order)
    plt.title(f"Distribución de {col}")
    plt.xlabel("Cantidad")
    plt.ylabel(col)
    plt.show()
```
